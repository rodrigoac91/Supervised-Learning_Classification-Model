Import the following packages:

```python
# importing libraries
import pandas as pd # data science essentials
import matplotlib.pyplot as plt # essential graphical output
import seaborn as sns # enhanced graphical output
import statsmodels.formula.api as smf # predictive modeling with nice outputs
import numpy as np 
from sklearn.model_selection import train_test_split # train/test split
from sklearn.linear_model import LinearRegression # linear regression (scikit-learn)
from sklearn.neighbors import KNeighborsRegressor # KNN for Regression
from sklearn.preprocessing import StandardScaler # standard scaler


# setting pandas print options
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)


# specifying file name
file = './birthweight_low.xlsx'


# reading the file into Python
df = pd.read_excel(io=file)


# outputting the first ten rows of the dataset
df.head(n=5)
```

Delete columns with NaN

```python
#Delete columns with no data and that are not to be used for regression

df1= df.drop(['Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20', 'Unnamed: 21', 'Unnamed: 22',
         'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25', 'Unnamed: 26', 'Unnamed: 27',
         'Unnamed: 28', 'Unnamed: 29', 'Unnamed: 30', 'Unnamed: 31', 'Unnamed: 32', 
         'Unnamed: 33'], axis=1)

df1.head(n=5)
```

Changin missing values 

```python
#Changing missing values by the median

meduc_med = df1 ['meduc'].median()
df1['meduc'].fillna(value = meduc_med,
                       inplace = True)

npvis_med = df1 ['npvis'].median()
df1['npvis'].fillna(value = npvis_med,
                       inplace = True)


feduc_med = df1 ['feduc'].median()
df1['feduc'].fillna(value = feduc_med,
                       inplace = True)
```

```python
#Check for null values
df1.isnull().any()
```

Adding a new column with one variable

```python
#Adding a new column
df1['new_npvis'] = df['npvis']*df['npvis']
```

```python
#Verify that the new column is added correctly
df1.head(n=5)
```

```python
#check for a null value
df1.isnull().any()
```

```python
#Insert median at null value

meduc_new_npvis = df1 ['new_npvis'].median()
df1['new_npvis'].fillna(value = meduc_new_npvis,
                       inplace = True)
```

```python
#check for a null value
df1.isnull().any()
```

```python
df1.info()
```

```python
#Filtering fage values
df1['fage'] = np.where(df1['fage']<40, 99997, df1['fage'])
df1['fage'] = np.where(df1['fage']<99, 99996, df1['fage'])
df1['fage'] = df1['fage'].astype('str')
df1=df1.replace('99997', 'Lowerthan40')
df1=df1.replace('99996', 'Higherthan40')

print(df1.value_counts(['fage']))
```

```python
#Verifying the application of the filters in the column

data_features= df1.columns

tr_f = df1[data_features]

tr_f.head(n=5)
```

```python
#Applying dummies inside the table
df2 = pd.get_dummies(tr_f)


df2.head(n=5)
```

Applying logarithm in variable for then apply correlation

```python
#Applying logarithm in variables 
df2['log_meduc'] = np.log(df2['meduc'])
df2['log_monpre']   = np.log(df2['monpre'])
df2['log_feduc'] = np.log(df2['feduc'])
df2['log_new_npvis'] = np.log(df2['new_npvis'])
df2['log_bwght'] = np.log(df2['bwght'])
df2['log_npvis'] = np.log(df2['npvis'])
df2['log_mage']   = np.log(df2['mage'])

#sum +0.0009 because there are variable 0
df2['log_cigs']   = np.log(df2['cigs'] + 0.0009)
df2['log_drink']   = np.log(df2['drink'] + 0.0009)
```

```python
for column in df2:
    print(f"{column}")
```

Creating a continuous data with the variable for correlation

```python
# creating a list of continuous features (including 'bwght')
continuous_data_log = [ 'bwght', 
                        'meduc',
                        'monpre',
                        'npvis',
                        'feduc',
                        'cigs',
                        'drink',
                        'male',
                        'mwhte',
                        'mblck',
                        'moth',
                        'fwhte',
                        'fblck',
                        'foth',
                        'new_npvis',
                        'fage_Higherthan40',
                        'fage_Lowerthan40',
                        'log_mage',
                        'log_meduc',
                        'log_monpre',
                        'log_feduc',
                        'log_new_npvis',
                        'log_npvis',
                        'log_cigs',
                        'log_drink']
                  

# developing a correlation matrix based on continuous features
df2_corr = df2[continuous_data_log].corr(method = 'pearson')


# filtering the results to only show correlations with bwght
df2_corr.loc[ : , continuous_data_log].round(decimals = 2).sort_values(ascending = False,
                                                                      by="bwght")
```

Building a base model

```python
# Step 1: build a model
lm_best = smf.ols(formula = """bwght ~  cigs+
                                        drink +
                                        npvis +
                                        log_mage +
                                        fage_Higherthan40 +
                                        log_new_npvis""",
                                        data = df2)


# Step 2: FIT the data into the model object
results = lm_best.fit()


# Step 3: analyze the SUMMARY output
print(results.summary())
```
